#Importing my_data_apple

```python
def bad_line(x):
    print(x)
    return None

df_apple = pd.read_csv('/content/drive/Othercomputers/My Laptop (1)/year4/final_project/data_science/data_set/my_data_apple.csv',  on_bad_lines=bad_line, engine='python')
```

```python
# drop column 
df_apple = df_apple.drop(columns=['Tweet_id','Cordinates','Place','User'])
df_apple.dtypes
```

Cleaning Date column


```python
# drop empty and non-date values in the 'date' column
df_apple['Date'] = pd.to_datetime(df_apple['Date'], errors='coerce').dt.tz_localize(None)
df_apple = df_apple.dropna(subset=['Date'])
df_apple['Tweet'] = df_apple['Tweet'].astype('string')
column_list = ['Hashtags','Cashtags','Language'] 

# replace None and NaN with null in column 'Cashtags'
for column_item in column_list:
  df_apple[column_item] = df_apple[column_item].fillna('null')
  df_apple[column_item] = df_apple[column_item].astype('string')

drop_list = ['Likes','Comments','Retweets','Views']
for drop_item in drop_list:
  df_apple = df_apple.dropna(subset=[drop_item])
```

```python
# replace empty strings and whitespace with NaN
df_apple['Mentioned_users'] = df_apple['Mentioned_users'].replace(['', ' '], np.nan)

# convert to bool
df_apple['Mentioned_users'] = df_apple['Mentioned_users'].notnull().astype(bool)

# convert column 'Views' to numeric data type
df_apple['Views'] = pd.to_numeric(df_apple['Views'], errors='coerce')
df_apple['User_created_date'] = pd.to_datetime(df_apple['User_created_date'], errors='coerce').dt.tz_localize(None)
df_apple['Favourite_count'] = pd.to_numeric(df_apple['Favourite_count'], errors='coerce')

# replace empty strings and whitespace with NaN
df_apple['Location'] = df_apple['Location'].replace(['', ' '], np.nan)

# convert to bool
df_apple['Location'] = df_apple['Location'].notnull().astype(bool)
df_apple['Verified'] = df_apple['Verified'].notnull().astype(bool)
df_apple['Average_favourite_count'] = df_apple['Favourite_count'] / df_apple['Tweet_count']
# calculate account age in years
current_date = dt.datetime.now()
df_apple['account_age'] = (current_date - df_apple['User_created_date']).astype('timedelta64[Y]')

# extract the time of day in 24 hour format
df_apple['time'] = df_apple['Date'].dt.strftime('%H:%M:%S')

# create a Day of week column
df_apple['Day of week'] = df_apple['Date'].dt.day_name()

#reorder
df_apple = df_apple.reindex(columns=['Tweet', 'Date', 'time', 'Day of week', 'Cashtags', 'Hashtags', 'Language', 'Location', 'Mentioned_users', 'Followers', 'Following', 'User_created_date', 'Listed_count', 'Favourite_count', 'Tweet_count', 'Verified', 'Average_favourite_count', 'account_age', 'Likes', 'Comments', 'Retweets', 'Views'])
```
