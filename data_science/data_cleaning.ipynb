Checking the data of Tweet_id


```python
print(df_apple.isna().sum())
```

```python

# drop empty and non-date values in the 'date' column
df_apple['Date'] = pd.to_datetime(df_apple['Date'], errors='coerce').dt.tz_localize(None)
```

```python
df_apple = df_apple.dropna(subset=['Date'])
```

```python
df_apple['Tweet'] = df_apple['Tweet'].astype('string')

# Convert column 'A' to string data type and drop rows with non-string values
# df_apple = df_apple[df_apple['Tweet'].apply(lambda x: type(x)==str)]
```

```python
df_apple.dtypes
```

```python

# replace None and NaN with null in column 'Cashtags'
df_apple['Cashtags'] = df_apple['Cashtags'].fillna('null')
df_apple['Cashtags'] = df_apple['Cashtags'].astype('string')
```

```python
print(df_apple.isna().sum())
```

```python
list = ['Hashtags','Cashtags','Language'] 
# replace None and NaN with null in column 'Cashtags'
for item in list:
  df_apple[item] = df_apple[item].fillna('null')
  df_apple[item] = df_apple[item].astype('string')
```
